# Assignment # 02 Add Columns

```python
import pandas as pd
data = pd.read_csv("Pokemon.csv")
data.head()
```

```python
speed_mean = data["Speed"].mean()
print(data["Speed"].mean())
def set_speed(val):
    if val > speed_mean:
        return "Speed High"
    elif val == speed_mean:
        return "Speed Normal"
    else:
        return "Speed Low"

data["Speed_high_low"] = data["Speed"].apply(set_speed)

hp_mean = data["HP"].mean()
print(data["HP"].mean())
def set_hp(val):
    if val > hp_mean:
        return "HP High"
    elif val == hp_mean:
        return "HP Normal"
    else:
        return "HP Low"

    # "HP_High" -> "HP High", "HP Low"
data["HP_high_low"] = data["HP"].apply(set_hp)
data
```
